# Imports

```python
import pandas as pd
import geopandas as gpd
from matplotlib import pyplot as plt
from mpl_toolkits import mplot3d
import json
```

# Plotting Function

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

def plot_trajectory(traj,cluster=-1,name=""):
    
    if len(traj[0]) == 2:
        l1,l2 = map(list, zip(*traj))
        plt.plot(l1,l2,c="C" + str(cluster)) if cluster>=0 else plt.plot(l1,l2)
        #plt.savefig(name + ".png", bbox_inches='tight')
        return
    elif len(traj[0]) == 3:
        l1,l2,l3 = map(list,zip(*traj))
        ax.plot3D(l1,l2,l3,c="C" + str(cluster)) if cluster>=0 else ax.plot3D(l1,l2,l3)
        #plt.savefig(name + ".png", bbox_inches='tight')
        return
    else:
        raise ValueError("can only handle 2D or 3D trajectories")
```

# Sample Visualization

```python
sample = pd.read_csv("datasets/sample.csv",sep=";")
sample.apply(lambda x: plot_trajectory(json.loads(x["trajectory"]),x["symbol_id"],"Sample"),axis=1)
plt.show()
```

Above you should see a visualization of a snippet of the DigiLeTs dataset. Colors symbolise ground truth clusters (which in this case refer to individual symbols)
